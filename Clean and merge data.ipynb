```python
# Creating two new columns
# Function to extract numeric part and convert to int
def convert_to_int(value):
    try:
        return int(''.join(filter(str.isdigit, str(value))))
    except ValueError:
        return None  # Handle non-convertible values

# Apply the conversion function to the column
epl_df['Minutes'] = epl_df['Minutes'].apply(convert_to_int)
epl_df['Mins_Per_Match'] = (epl_df['Minutes']/epl_df['Match Played']).astype(int)
epl_df['Goals_Per_Match'] = (epl_df['Goals']/epl_df['Match Played']).astype(float)
epl_df.head()
```
